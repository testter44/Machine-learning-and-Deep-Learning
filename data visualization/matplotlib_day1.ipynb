<h3 style="color:blue;">Line Charts</h3> 

* represents information as a series of data points connected by a straight line
* each data point or marker is plotted and connected with a line or curve

```python
import matplotlib.pyplot as plt
```

```python
# Sample data
months = ['January', 'February', 'March', 'April', 'May', 'June']
sales = [100, 120, 90, 80, 110, 130]

# Creating the line plot
plt.plot(months, sales, marker='o', linestyle='-', color='y')

# Customizing the plot
plt.title('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid(True)

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
months = ['January', 'February', 'March', 'April', 'May', 'June']
sales_A = [100, 120, 90, 80, 110, 130]
sales_B = [80, 110, 100, 90, 120, 100]
sales_C = [90, 100, 110, 120, 130, 140]

# Changing the figure size
plt.figure(figsize=(12, 6))  # Adjust the width and height as desired

# Creating the line plots
plt.plot(months, sales_A, marker='d', linestyle='-', color='b', label='Sales A')
plt.plot(months, sales_B, marker='s', linestyle='-', color='r', label='Sales B')
plt.plot(months, sales_C, marker='1', linestyle=':', color='g', label='Sales C')

# Customizing the plot
plt.title('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid(True)
plt.legend()  # add legend to plot 

# Displaying the plot
plt.show()
```

<h3 style="color:blue;">Bar Graph</h3> 

* categorical data can be represented in a bar graph
* plots data with the help of bars, which represent value on the y-axis and category on the x-axis

```python
import matplotlib.pyplot as plt

# Sample data
section = ['section A', 'section B', 'section C', 'section D', 'a', 'b']
number_of_students = [25, 40, 30, 35, 67, 54]

# Creating the bar graph
plt.bar(section, number_of_students)

# Customizing the plot
plt.title('Bar Graph showing students with their section')
plt.xlabel('section')
plt.ylabel('number_of_students')

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
section = ['section A', 'section B', 'section C', 'section D']
number_of_boys_students = [25, 40, 30, 35]
number_of_girls_students = [15, 20, 25, 30]

# Creating the stacked bar graph
plt.bar(section, number_of_boys_students, label='number of boys students')
plt.bar(section, number_of_girls_students, bottom=number_of_girls_students, label='number of girls students')

# Customizing the plot
plt.title('Bar Graph showing students with their section')
plt.xlabel('section')
plt.ylabel('number of students')
plt.legend()

# Displaying the plot
plt.show()
```

<h3 style="color:blue;">Scatter Plots</h3> 

* Scatter plots are used when we have to plot two or more variables present at different coordinates
* The data is scattered all over the graph and is not confined to a range

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)

# Customizing the plot
plt.title('Scatter Plot')
plt.xlabel('Age of car')
plt.ylabel('speed of the car')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data
np.random.seed(42)  # Set seed for reproducibility
num_students = 100

# Class A (Male students)
x_a = np.random.uniform(0, 10, size=num_students // 2)
y_a = 3 * x_a + np.random.normal(0, 2, size=num_students // 2)
# print(y_a)

# Class A (Female students)
x_b = np.random.uniform(0, 10, size=num_students // 2)
y_b = 5 * x_b + np.random.normal(0, 1.5, size=num_students // 2)

# Creating the scatter plot
plt.scatter(x_a, y_a, color='blue', label='Male')
plt.scatter(x_b, y_b, color='red', label='Female')

# Customizing the plot
plt.title('Scatter Plot of Students')
plt.xlabel('Study Hours')
plt.ylabel('Test Scores')
plt.legend()

# Displaying the plot
plt.show()
```

<h3 style="color:blue;">Heat Maps</h3> 

* used to see changes in behavior or gradual changes in data
* visualizes the value of the matrix with colours
*  brighter the colour means the higher the value and vice versa 

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.random((12, 12))

plt.imshow(data, cmap='autumn', interpolation='nearest')
 
# Add colorbar
plt.colorbar()
 
plt.title("Heatmap with color bar")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
cities = ['City A', 'City B', 'City C', 'City D']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
temperature_data = np.random.randint(20, 40, size=(len(cities), len(months)))

# Create heatmap
fig, ax = plt.subplots()
im = ax.imshow(temperature_data, cmap='YlOrRd')

# Customize the plot
ax.set_xticks(np.arange(len(months)))
ax.set_yticks(np.arange(len(cities)))
ax.set_xticklabels(months)
ax.set_yticklabels(cities)
plt.xlabel('Months')
plt.ylabel('Cities')
plt.title('Temperature Variation Heatmap')

# Add colorbar
cbar = ax.figure.colorbar(im, ax=ax, shrink=0.6)
cbar.ax.set_ylabel('Temperature', rotation=-90, va='bottom')

# Display the heatmap
plt.show()
```
